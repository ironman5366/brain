```python
import matplotlib.pyplot as plt

def visualize_sample(sample, times=None, ch_names=None, figsize=(14, 20)):
    """
    Visualize an EEG sample as separate graphs for each channel.
    
    Args:
        sample: Array of shape (Channels, Samples), e.g. (32, 250)
        times: Optional array of time points (0-1 normalized). Will be rescaled to -200ms to 1000ms.
        ch_names: Optional list of channel names. If None, uses "Channel N".
        figsize: Figure size as (width, height) tuple.
    """
    n_channels, n_samples = sample.shape
    
    if times is None:
        times = np.arange(n_samples)
        x_label = "Sample"
    else:
        # Rescale from 0-1 to -200ms to 1000ms
        times = -200 + times * 1200
        x_label = "Time (ms)"
    
    if ch_names is None:
        ch_names = [f"Channel {i}" for i in range(n_channels)]
    
    fig, axes = plt.subplots(n_channels, 1, figsize=figsize, sharex=True)
    
    for i, (ax, name) in enumerate(zip(axes, ch_names)):
        ax.plot(times, sample[i], linewidth=0.8)
        ax.set_ylabel(name, fontsize=8, rotation=0, labelpad=40, ha='right')
        ax.tick_params(axis='y', labelsize=6)
        if i < n_channels - 1:
            ax.tick_params(axis='x', labelbottom=False)
    
    # Add vertical lines at key time points
    for ax in axes:
        ax.axvline(x=0, color='red', linestyle='--', linewidth=0.5, alpha=0.7)      # Stimulus onset
        ax.axvline(x=100, color='green', linestyle='--', linewidth=0.5, alpha=0.7)  # 100ms
        ax.axvline(x=200, color='blue', linestyle='--', linewidth=0.5, alpha=0.7)   # 200ms
    
    axes[-1].set_xlabel(x_label)
    plt.tight_layout()
    plt.show()
    
# Example usage:
# visualize_sample(t["preprocessed_eeg_data"][0], times=t["times"], ch_names=t["ch_names"])
```

```python
def visualize_sequence(data, channel=0, times=None, ch_names=None, figsize=(16, 4)):
    """
    Visualize multiple sequences for a single channel, concatenated one after another.
    
    Args:
        data: Array of shape (N, Channels, Samples), e.g. (10, 32, 250)
        channel: Which channel to visualize (index or name if ch_names provided).
        times: Optional array of time points (0-1 normalized). Will be rescaled to -200ms to 1000ms.
        ch_names: Optional list of channel names.
        figsize: Figure size as (width, height) tuple.
    """
    n_sequences, n_channels, n_samples = data.shape
    
    # Resolve channel index if name provided
    ch_idx = channel
    if isinstance(channel, str) and ch_names is not None:
        ch_idx = list(ch_names).index(channel)
    
    ch_label = ch_names[ch_idx] if ch_names is not None else f"Channel {ch_idx}"
    
    # Concatenate all sequences for the selected channel
    concatenated = data[:, ch_idx, :].flatten()
    
    # Create x-axis
    if times is not None:
        # Rescale from 0-1 to -200ms to 1000ms per sequence
        single_times = -200 + times * 1200
        x_vals = np.concatenate([single_times + i * 1200 for i in range(n_sequences)])
        x_label = "Time (ms)"
    else:
        x_vals = np.arange(len(concatenated))
        x_label = "Sample"
    
    fig, ax = plt.subplots(figsize=figsize)
    ax.plot(x_vals, concatenated, linewidth=0.8)
    
    # Add vertical lines at sequence boundaries
    for i in range(1, n_sequences):
        if times is not None:
            boundary = -200 + i * 1200
        else:
            boundary = i * n_samples
        ax.axvline(x=boundary, color='red', linestyle='--', linewidth=1, alpha=0.7)
    
    ax.set_xlabel(x_label)
    ax.set_ylabel(ch_label)
    ax.set_title(f"{ch_label} across {n_sequences} sequences")
    plt.tight_layout()
    plt.show()

# Example usage:
# visualize_sequence(t["preprocessed_eeg_data"][:10], channel=0, times=t["times"], ch_names=t["ch_names"])
```

```python
def visualize(data, lines=None, ch_names=None, figsize=(14, 20), line_color='red'):
    """
    Visualize arbitrary EEG data as separate graphs for each channel.
    
    Args:
        data: Array of shape (Channels, Samples)
        lines: Optional list of x positions to draw vertical lines, e.g. [100, 200]
        ch_names: Optional list of channel names. If None, uses "Channel N".
        figsize: Figure size as (width, height) tuple.
        line_color: Color for vertical lines (default 'red').
    """
    n_channels, n_samples = data.shape
    x_vals = np.arange(n_samples)
    
    if ch_names is None:
        ch_names = [f"Channel {i}" for i in range(n_channels)]
    
    fig, axes = plt.subplots(n_channels, 1, figsize=figsize, sharex=True)
    
    # Handle single channel case
    if n_channels == 1:
        axes = [axes]
    
    for i, (ax, name) in enumerate(zip(axes, ch_names)):
        ax.plot(x_vals, data[i], linewidth=0.8)
        ax.set_ylabel(name, fontsize=8, rotation=0, labelpad=40, ha='right')
        ax.tick_params(axis='y', labelsize=6)
        if i < n_channels - 1:
            ax.tick_params(axis='x', labelbottom=False)
    
    # Add vertical lines at specified positions
    if lines is not None:
        for line_pos in lines:
            for ax in axes:
                ax.axvline(x=line_pos, color=line_color, linestyle='--', linewidth=0.5, alpha=0.7)
    
    axes[-1].set_xlabel("Sample")
    plt.tight_layout()
    plt.show()

# Example usage:
# visualize(t["preprocessed_eeg_data"][0], lines=[100, 200], ch_names=t["ch_names"])
```
