```python
from safetensors.torch import load_file
```

```python
from pathlib import Path
DS_NAME = "things-processed-fixed-v2-2025-12-26"
THINGS_BASE = Path(
    "/kreka/research/willy/side/brain_datasets/things/"
) / DS_NAME
```

```python
data_file = THINGS_BASE / f"{DS_NAME}.safetensors"
metadata_file = THINGS_BASE / f"{DS_NAME}-metadata.parquet"
```

```python
samples = load_file(data_file)["samples"]
samples.shape, samples
```

```python
import polars as pl
df = pl.read_parquet(metadata_file).with_row_index()
df
```

```python
objects = df.unique("object")
objects
```

```python
fifty_objects = objects["object"][0:50]
fifty_objects
```

```python
obj_filtered = df.filter(
    pl.col("object").is_in(fifty_objects)
)
obj_filtered
```

```python
import torch

obj_mask = torch.zeros(
    samples.shape[0],
    dtype=torch.bool
)

for row in obj_filtered.iter_rows(named=True):
    obj_mask[row["index"]] = True

obj_mask.nonzero().shape
```

```python
selected_samples = samples[obj_mask]
selected_samples.shape
```

```python
OUT_PATH = Path("/kreka/research/willy/side/brain_datasets/things/") / "fifty_objects"
OUT_PATH.mkdir(
    exist_ok=True,
    parents=False
)
```

```python
from safetensors.torch import save_file

assert len(obj_filtered) == len(selected_samples), "Something went wrong"

out_name = "fifty_objects"
data_path = OUT_PATH / f"{out_name}.safetensors"
metadata_path = OUT_PATH / f"{out_name}-metadata.parquet"

save_file(
    {
        "samples": selected_samples
    },
    data_path
)
obj_filtered.write_parquet(
    metadata_path
)

print(f"Saved data to {data_path}, metadata to {metadata_path}")
```
