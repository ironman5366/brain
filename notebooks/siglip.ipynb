```python
from concurrent.futures import ThreadPoolExecutor
from tqdm import tqdm
from PIL import Image
import traceback

def load_image(im_id):
    try:
        im_path = STIMULI_PATH / f"{im_id:05d}.jpg"
        i = Image.open(im_path).convert('RGB')
        return im_id, i
    except Exception as e:
        # Print so threading doesn't swallow the exception
        traceback.print_exc()
        print("err", im_id, e)
        raise e
    

def load_all_images(image_df: pl.DataFrame):
    loaded_images = {}


    with ThreadPoolExecutor(32) as executor:
        futs = []
        for row in tqdm(image_df.iter_rows(named=True), desc="Submitting..."):
            futs.append(
                executor.submit(
                    load_image,
                    row["image_id"]
                )
            )
        
        for fut in tqdm(
            futs,
            desc="Loading..."
        ):
            image_id, im = fut.result()
            loaded_images[image_id] = im
    
    return loaded_images
```

```python
loaded_images = load_all_images(unique_images)
```

```python
DEVICE = "cuda"
```

```python
from transformers import AutoModel, AutoProcessor

model = AutoModel.from_pretrained("google/siglip2-so400m-patch16-naflex").to(DEVICE)
processor = AutoProcessor.from_pretrained(
    "google/siglip2-so400m-patch16-naflex", use_fast=True
)
```

```python
first_ims = list(loaded_images.values())[:5]
first_ims
```

```python
first_inputs = processor(images=first_ims, return_tensors="pt")
first_inputs = {
    k: v.to(DEVICE)
    for k,v in first_inputs.items()
}
model.get_image_features(**first_inputs).shape
```

```python
from tqdm import tqdm
import torch
import gc

BATCH_SIZE = 384
model = model.to(DEVICE)


def embed_images(ims):
    with torch.inference_mode():
        all_embeds = []    
        for i in tqdm(range(0, len(ims), BATCH_SIZE), desc="Embedding images"):
            gc.collect()
            torch.cuda.empty_cache()

            batch = ims[i:i + BATCH_SIZE]
            print(f"Processing batch of {len(batch)}")
            inputs = processor(images=batch, return_tensors="pt")
            inputs = {k: v.to(DEVICE) for k, v in inputs.items()}
            embeds = model.get_image_features(**inputs)
            del inputs
            all_embeds.append(embeds)
            
        
        return torch.cat(all_embeds, dim=0)
```

```python
rearranged_keys = list(sorted(
    loaded_images.keys()  
))
```

```python
rearranged_loaded = {k: loaded_images[k] for k in rearranged_keys}
```

```python
embeds = embed_images(list(rearranged_loaded.values()))
```

```python
import random
rand = random.choice(unique_images)
print(rand)
image_id = rand["image_id"][0]
computed_emb = embeds[image_id]
print("image id", rand["image_id"])
_, loaded_im = load_image(image_id)
with torch.inference_mode(): 
    inputs = processor(images=[loaded_im], return_tensors="pt")
    inputs = {k: v.to(model.device) for k, v in inputs.items()}
    sanity_embed = model.get_image_features(**inputs)[0].to(computed_emb.dtype)
    assert torch.allclose(sanity_embed.to(computed_emb.device), computed_emb, atol=1e-4)
```
